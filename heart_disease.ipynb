```python
df['TenYearCHD'].value_counts()
```

```python
sns.countplot(x='TenYearCHD',data=df)
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class 
df_majority = df[(df['TenYearCHD']==0)] 
df_minority = df[(df['TenYearCHD']==1)] 
#upsample minority class
df_minority_upsampled = resample(df_minority, replace=True,n_samples= 511,random_state=42)
#Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
df_upsampled['TenYearCHD'].value_counts()
```

```python
df=df_upsampled
```

```python
sns.countplot(x='TenYearCHD',data=df)
```

```python
#splitting the features and target
```

```python
X=df.drop(columns='TenYearCHD',axis=1)
y=df['TenYearCHD']
```

```python
#standardizing the input varables
from sklearn.preprocessing import StandardScaler 
```

```python
X_std=StandardScaler().fit_transform(X)
```

```python
#Splitting the Data into Training data & Test Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_std,y,test_size=0.2,stratify=y,random_state=0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
#Model Training
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
#Training the LogisticRegression model with Training data
```

```python
model.fit(X_train,y_train)
```

```python
#model evaluation
```

```python
#Accuracy score
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Accuracy on Training data
```

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
#Accuracy on Test data
```

```python
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
```

```python
print('Accuray on Test data:',test_data_accuracy)
```

```python
#Building a Predictive System
```

```python
input_data=(0,66,0,0,0,1,0,214,212,25.32,57,84)

#change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)

print(prediction)

if(prediction[0] == 1):
    print("The person has heart disease")
else:
    print("The person does not have heart disease")
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
sns.heatmap(cm,linewidth=.01,annot=True,cmap="winter",annot_kws={"size": 20})
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]

print('Testing Accuracy for Logistic Regression:',(TP+TN)/(TP+TN+FN+FP))
print('Testing Sensitivity for Logistic Regression:',(TP/(TP+FN)))
print('Testing Specificity for Logistic Regression:',(TN/(TN+FP)))
print('Testing Precision for Logistic Regression:',(TP/(TP+FP)))
```

```python
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=500,criterion='entropy',max_depth=8,min_samples_split=5)
model3 = rfc.fit(X_train, y_train)
prediction3 = model3.predict(X_test)
cm3=confusion_matrix(y_test, prediction3)
sns.heatmap(cm3, annot=True,cmap='winter',linewidths=0.3, linecolor='black',annot_kws={"size": 20})
TP=cm3[0][0]
TN=cm3[1][1]
FN=cm3[1][0]
FP=cm3[0][1]
print(round(accuracy_score(prediction3,y_test)*100,2))
print('Testing Accuracy for Random Forest:',(TP+TN)/(TP+TN+FN+FP))
print('Testing Sensitivity for Random Forest:',(TP/(TP+FN)))
print('Testing Specificity for Random Forest:',(TN/(TN+FP)))
print('Testing Precision for Random Forest:',(TP/(TP+FP)))
```

```python
cm3
```

```python
print(classification_report(y_test,prediction3))
```

```python
from sklearn.svm import SVC
svm=SVC(C=12,kernel='linear')
model4=svm.fit(X_train,y_train)
prediction4=model4.predict(X_test)
cm4= confusion_matrix(y_test,prediction4)
sns.heatmap(cm4, annot=True,cmap='winter',linewidths=0.3, linecolor='black',annot_kws={"size": 20})
TP=cm4[0][0]
TN=cm4[1][1]
FN=cm4[1][0]
FP=cm4[0][1]
  
print('Testing Accuracy for SVM:',(TP+TN)/(TP+TN+FN+FP))
print('Testing Sensitivity for Random Forest:',(TP/(TP+FN)))
print('Testing Specificity for Random Forest:',(TN/(TN+FP)))
print('Testing Precision for Random Forest:',(TP/(TP+FP)))
```
